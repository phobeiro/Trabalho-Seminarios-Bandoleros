```python
import pandas as pd
import os
import warnings
import glob
import seaborn as sns


def junta_tudo(folder, colunas=None):
    
    df = pd.DataFrame()
    caminho_pasta = os.path.join(folder, '*csv')
    files = glob.glob(caminho_pasta)
    for file in files:
        dfloop = pd.read_csv(file, sep=';', usecols=colunas)
        df = pd.concat([df,dfloop])
    return df
```

```python
infracoesAll = open('infracoes_all.txt','r')
infracoesSim = open('infracoes_s.txt', 'r')

todasInfra = []
simInfra = []

for l in infracoesAll:
    linha = l.strip()
    todasInfra.append(linha)

for l in infracoesSim:
    linha = l.strip()
    simInfra.append(linha)
```

```python
frame = pd.DataFrame(todasInfra, columns=['Todas'])

frame.loc[frame.Todas.isin(simInfra), 'Selecionada'] = 'Sim'
```

```python
frame.fillna('Não', inplace=True)
frame.sort_values(by=['Selecionada'], inplace=True, ascending=False)
frame.head()
frame.to_csv('infra_select_sorted.csv', index=False, sep=';')
```
